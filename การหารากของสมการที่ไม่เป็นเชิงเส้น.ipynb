```python
import numpy as np
import matplotlib.pyplot as plt

def f(x):
    return x**3 - x - 2  # ฟังก์ชันที่เราต้องการหา root

# กำหนดช่วงของ x
x = np.linspace(-3, 3, 400)
y = f(x)

plt.plot(x, y, label='f(x) = x^3 - x - 2')
plt.axhline(0, color='black',linewidth=0.5)  # แกน x
plt.axvline(0, color='black',linewidth=0.5)  # แกน y
plt.grid(color = 'gray', linestyle = '--', linewidth = 0.5)
plt.legend()
plt.title('Graphical Method')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.show()
```

```python
def g(x):
    return (x + 2) / 3  # การเปลี่ยนสมการ x^3 - x - 2 = 0 เป็น x = (x + 2) / 3

def fixed_point_iteration(x0, tol=1e-6, max_iter=100):
    x = x0
    for i in range(max_iter):
        x_new = g(x)
        if abs(x_new - x) < tol:  # ตรวจสอบความแม่นยำ
            return x_new
        x = x_new
    raise ValueError("Solution did not converge")

root = fixed_point_iteration(1.0)  # เริ่มต้นที่ 1.0
print(f'Root: {root}')
```

```python
def f(x):
    return x**3 - x - 2

def bisection_method(a, b, tol=1e-6, max_iter=100):
    if f(a) * f(b) >= 0:
        raise ValueError("The function must have different signs at endpoints a and b.")
    for _ in range(max_iter):
        c = (a + b) / 2
        if abs(f(c)) < tol:
            return c
        elif f(a) * f(c) < 0:
            b = c
        else:
            a = c
    raise ValueError("Solution did not converge")

root = bisection_method(1, 2)
print(f'Root: {root}')
```

```python
def f(x):
    return x**3 - x - 2

def regula_falsi_method(a, b, tol=1e-6, max_iter=100):
    if f(a) * f(b) >= 0:
        raise ValueError("The function must have different signs at endpoints a and b.")
    for _ in range(max_iter):
        c = b - (f(b) * (a - b)) / (f(a) - f(b))
        if abs(f(c)) < tol:
            return c
        elif f(a) * f(c) < 0:
            b = c
        else:
            a = c
    raise ValueError("Solution did not converge")

root = regula_falsi_method(1, 2)
print(f'Root: {root}')
```

```python
def f(x):
    return x**3 - x - 2

def secant_method(x0, x1, tol=1e-6, max_iter=100):
    for _ in range(max_iter):
        f0 = f(x0)
        f1 = f(x1)
        if f1 - f0 == 0:
            raise ValueError("Zero denominator in secant method")
        x2 = x1 - f1 * (x1 - x0) / (f1 - f0)
        if abs(f(x2)) < tol:
            return x2
        x0, x1 = x1, x2
    raise ValueError("Solution did not converge")

root = secant_method(1, 2)
print(f'Root: {root}')
```

```python
from scipy.optimize import newton

def f(x):
    return x**3 - x - 2

def df(x):
    return 3*x**2 - 1

root = newton(f, x0=1.5, fprime=df)
print(f'Root: {root}')
```
